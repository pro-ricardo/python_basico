# Exemplo de uma aplicação em Python: NUVEM DE PALAVRAS

```python
import pandas as pd #Biblioteca responsável para fazer a leitura e impressão de bases de dados
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
```

```python
filmes=pd.read_csv('https://raw.githubusercontent.com/profandersonvanin01/python_introducao/main/netflix_titles.csv')
```

```python
filmes = pd.DataFrame(filmes)
filmes.head()
```

```python
filmes.shape
```

Nesse exemplo vamos selecionar apenas os filmes brasileiros e em seguida, selecionar a coluna que vamos usar para formar a wordcloud.

```python
filmes_brasil = filmes[filmes['country'] == 'Brazil']
```

```python
palavras = filmes_brasil['description']
```

```python
todas_palavras = " ".join(w for w in palavras)
```

```python
todas_palavras
```

```python
stopwords = set(STOPWORDS)
```

A princípio, vamos definir uma função que irá plotar nossa wordcloud. Em seguida, vamos usar a biblioteca wordcloud para gerar nossa nuvem de palavras.

```python
def imprime_nuvem(wc):
  fig, ax = plt.subplots(figsize=(14,7))
  ax.imshow(wc, interpolation='bilinear')
  ax.set_axis_off()
  plt.imshow(wc)
```

```python
wc = WordCloud(stopwords=stopwords,
               background_color='white',
               width=1600,
               height=800).generate(todas_palavras)
```

```python
imprime_nuvem(wc)
```

Agora vamos fazer plotar novamente essa wordcloud escolhendo as cores que queremos visualizar

```python
cores = ['darkgreen','navy','gold']
my_cmap = ListedColormap(sns.color_palette(cores).as_hex())
```

```python
wc_novo = WordCloud(stopwords=stopwords,
               colormap=my_cmap,
               background_color='white',
               width=1600,
               height=800).generate(todas_palavras)
```

```python
imprime_nuvem(wc_novo)
```

Para deixar nossa wordcloud ainda mais característica, vamos inseri-lo em uma imagem, que nesse caso será o mapa do Brasil. Dessa maneira, conseguimos customizar o formato e as cores utilizadas.

```python
brasil_mask = np.array(Image.open("/content/mapa_brasil.jpg"))
```

```python
image_colors = ImageColorGenerator(brasil_mask)
```

```python
wc_brasil = WordCloud(stopwords=stopwords,
                      max_font_size=100,
                      min_font_size=1,
                      color_func=image_colors,
                      mask=brasil_mask,
                      background_color='white',
                      width=1600,
                      height=800).generate(todas_palavras)
```

```python
imprime_nuvem(wc_brasil)
```
